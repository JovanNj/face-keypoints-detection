## LeNet

```python
class LeNet(nn.Module):
    def __init__(self, image_size=(1,224,224), output_size=136, kernels=[5,5], out_channels=[64, 128]):
        super().__init__()
        
        self.conv1 = nn.Conv2d(1, out_channels[0], kernels[0])
        self.conv2 = nn.Conv2d(out_channels[0], out_channels[1], kernels[1])
        
        self.pool = nn.MaxPool2d(2,2)
        
        n_size = self._get_conv_output(image_size)
        
        self.fc1 = nn.Linear(n_size, 544)
        self.fc2 = nn.Linear(544, 272)
        self.fc3 = nn.Linear(272, 136)
        
    def _get_conv_output(self, shape):
        bc = 1
        input = Variable(torch.rand(bc, *shape))
        output = self._forward_feat(input)
        n_size = output.data.view(bc, -1).size(1)
        return n_size
    
    def _forward_feat(self, x):
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        return x
      
        
    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        return x
        
        x = x.view(x.size(0), -1)
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.fc3(x)
        
```

## Naimi≈° mre≈æa

```python
class NaimisNet(nn.Module):
    def __init__(self, image_size=(1, 224, 224), output_size = 136, 
                 kernels = [5,5,5,5], out_channels = [32, 64, 128, 256], dropout=0.1):
        super().__init__()
        
        self.conv1 = nn.Conv2d(1, out_channels[0], kernels[0])
        self.conv2 = nn.Conv2d(out_channels[0], out_channels[1], kernels[1])
        self.conv3 = nn.Conv2d(out_channels[1], out_channels[2], kernels[2])
        self.conv4 = nn.Conv2d(out_channels[2], out_channels[3], kernels[3])
        
        self.drop = nn.Dropout(p=dropout)
        
        self.pool = nn.MaxPool2d(2, 2)
        
        n_size = self._get_conv_output(image_size)
        self.fc1 = nn.Linear(n_size, 1024)
        self.fc2 = nn.Linear(1024, 1024)
        self.fc3 = nn.Linear(1024, output_size)
        
    def _get_conv_output(self, shape):
        bc = 1
        input = Variable(torch.rand(bc, *shape))
        output_feat = self._forward_features(input)
        n_size = output_feat.data.view(bc, -1).size(1)
        return n_size
    
    def _forward_features(self, x):
        x = self.drop(self.pool(F.relu(self.conv1(x))))
        x = self.drop(self.pool(F.relu(self.conv2(x))))
        x = self.drop(self.pool(F.relu(self.conv3(x))))
        x = self.drop(self.pool(F.relu(self.conv4(x))))
        return x
    
    def forward(self, x):
        x = self.drop(self.pool(F.relu(self.conv1(x))))
        x = self.drop(self.pool(F.relu(self.conv2(x))))
        x = self.drop(self.pool(F.relu(self.conv3(x))))
        x = self.drop(self.pool(F.relu(self.conv4(x))))
        
        x = x.view(x.size(0), -1)
        x = self.drop(F.relu(self.fc1(x)))
        x = self.drop(F.relu(self.fc2(x)))
        x = self.fc3(x)
        
        return x
```

```python
from torch.utils.data.sampler import SubsetRandomSampler
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms, utils

from data_load import FacialKeypointsDataset
from data_load import Rescale, RandomRotate, RandomHorizontalFlip, RandomCrop, Normalize, ColorJitter, ToTensor, FaceCropTight, FaceCrop

def create_datasets(batch_size, img_size, color=False):
    valid_size = 0.5
    
    train_transform = transforms.Compose([RandomRotate(5),
                                          RandomHorizontalFlip(),
                                          ColorJitter(),
                                          FaceCrop(),
                                          Rescale((img_size,img_size)),
                                          Normalize(color=color),
                                          ToTensor()])
    valid_transform = transforms.Compose([FaceCropTight(),
                                          Rescale((img_size,img_size)),
                                          Normalize(color=color),
                                          ToTensor()])
    test_transform = transforms.Compose([FaceCropTight(),
                                          Rescale((img_size,img_size)),
                                          Normalize(color=color),
                                          ToTensor()])
                
    test_dataset = FacialKeypointsDataset(csv_file='data/test_frames_keypoints.csv',
                                          root_dir='data/test/',
                                          transform=test_transform)
    train_dataset = FacialKeypointsDataset(csv_file='data/training_frames_keypoints.csv',
                                           root_dir='data/training/',
                                           transform=train_transform)
    valid_dataset = FacialKeypointsDataset(csv_file='data/test_frames_keypoints.csv',
                                           root_dir='data/test/',
                                           transform=valid_transform)
            
    num_test = len(test_dataset)
    indices = list(range(num_test))
    np.random.shuffle(indices)
    split = int(np.floor(valid_size * num_test))
    test_idx, valid_idx = indices[split:], indices[:split]
   
    test_sampler = SubsetRandomSampler(test_idx)
    valid_sampler = SubsetRandomSampler(valid_idx)
    
    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
    valid_loader = DataLoader(valid_dataset, batch_size=batch_size, sampler=valid_sampler)
    test_loader = DataLoader(test_dataset, batch_size=batch_size, sampler=test_sampler)
    
    return train_loader, test_loader, valid_loader
```

```python
def val_loss(valid_loader, net):
    net.eval()
    net.to(device)
    loss, running_loss = 0.0, 0.0
    for i, batch in enumerate(valid_loader):
        images = batch['image']
        key_pts = batch['keypoints']
        
        key_pts = key_pts.view(key_pts.size(0), -1)
        
        images = images.type(torch.FloatTensor)
        key_pts = key_pts.type(torch.FloatTensor)
        
        images = images.to(device)
        key_pts = key_pts.to(device)
                
        output_pts = net(images)
        
        loss = criterion(output_pts, key_pts)
        running_loss += loss.item()
    avg_loss = running_loss/(i+1)
    net.train()
    return avg_loss
```

```python
class EarlyStopping:
    def __init__(self,patience=15):
        self.patience = patience
        self.counter = 0
        self.best_score = None
        self.early_stop = False
        self.val_loss_min = np.Inf
    
    def __call__(self, val_loss, model):
        score = -val_loss
        if self.best_score is None:
            self.best_score = score
        elif score < self.best_score:
            self.counter += 1
            print(f'EarlyStopping counter: {self.counter} out of {self.patience}')
            if self.counter >= self.patience:
                self.early_stop = True
        else:
            self.best_score = score
            self.save_checkpoint(val_loss, model)
            self.counter = 0
    
    def save_checkpoint(self, val_loss, model):
        '''Saves model when validation loss decrease.'''
        print(f'Validation loss decreased ({self.val_loss_min:.6f} --> {val_loss:.6f}).  Saving model ...')
        torch.save(model.state_dict(), 'saved_models/checkpoint.pt')
        self.val_loss_min = val_loss
```

```python
batch_size = 16
n_epochs = 3
train_loader, test_loader, valid_loader = create_datasets(batch_size,224)
train_loss_over_time = [] 
val_loss_over_time = [] 
early_stopping = EarlyStopping()
net = NaimisNet().to(device)

import torch.optim as optim
from torch.optim import lr_scheduler
device = torch.device('cuda')

criterion = nn.MSELoss()
optimizer = optim.Adam(params = net.parameters())

# reduce learning rate when a validation loss has stopped improving
scheduler = lr_scheduler.ReduceLROnPlateau(optimizer, 'min',  patience=7, verbose=True)
```

```python
net.train()
for epoch in range(n_epochs): 

    running_train_loss = 0.0
    avg_val_loss = 0.0
    avg_train_loss = 0.0

    for batch_i, data in enumerate(train_loader):
        images = data['image']
        key_pts = data['keypoints']

        key_pts = key_pts.view(key_pts.size(0), -1)

        key_pts = key_pts.type(torch.FloatTensor)
        images = images.type(torch.FloatTensor)

        images = images.to(device)
        key_pts = key_pts.to(device)

        output_pts = net(images)
        loss = criterion(output_pts, key_pts)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        running_train_loss += loss.item()
        if batch_i % 10 == 9 or batch_i == 0:    # print every 10 batches
            if batch_i == 0:
                avg_train_loss = running_train_loss
            else:
                avg_train_loss = running_train_loss/10
            avg_val_loss = val_loss(valid_loader, net)
            train_loss_over_time.append(avg_train_loss)
            val_loss_over_time.append(avg_val_loss)
            running_train_loss = 0.0

    scheduler.step(avg_val_loss)

    print(f'Epoch: {epoch + 1}, Avg. Training Loss: {avg_train_loss:.5f}, Avg. Validation Loss: {avg_val_loss:.5f}')
    early_stopping(avg_val_loss, net)

    if early_stopping.early_stop:
        train_loss_over_time = train_loss_over_time[:-early_stopping.patience]
        val_loss_over_time = val_loss_over_time[:-early_stopping.patience]
        print("Early stopping")
        break
```
