# Priprema podataka

## Uƒçitavanje i transformacija

```python
import numpy as np
import torch
import matplotlib.pyplot as plt
from torch.utils.data.sampler import SubsetRandomSampler
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms, utils

from data_load import FacialKeypointsDataset
from data_load import (Rescale, RandomRotate, RandomHorizontalFlip, RandomCrop, Normalize, ColorJitter,
                       ToTensor, FaceCrop, FaceCropTight)

device = torch.device('cuda')
```

```python
def create_datasets(batch_size, img_size, color=False):
    valid_size = 0.5
    # Transformacija podataka
    train_transform = transforms.Compose([RandomRotate(5),
                                          RandomHorizontalFlip(),
                                          ColorJitter(),
                                          FaceCrop(),
                                          Rescale((img_size, img_size)),
                                          Normalize(color=color),
                                          ToTensor()])
    
    valid_transform = transforms.Compose([FaceCropTight(),
                                          Rescale((img_size,img_size)),
                                          Normalize(color=color),
                                          ToTensor()])
    test_transform = transforms.Compose([FaceCropTight(),
                                          Rescale((img_size,img_size)),
                                          Normalize(color=color),
                                          ToTensor()])
    # Definisanje seta podataka
    test_dataset = FacialKeypointsDataset(csv_file='data/test_frames_keypoints.csv',
                                          root_dir='data/test/',
                                          transform=test_transform)

    train_dataset = FacialKeypointsDataset(csv_file='data/training_frames_keypoints.csv',
                                           root_dir='data/training/',
                                           transform=train_transform)
    
    valid_dataset = FacialKeypointsDataset(csv_file='data/test_frames_keypoints.csv',
                                           root_dir='data/test/',
                                           transform=valid_transform)
    
    # Kreiranje validacionog i test seta podataka
    n_test = len(test_dataset)
    idx = list(range(n_test))
    np.random.shuffle(idx)
    split = int(np.floor(valid_size * n_test))
    test_sampler = SubsetRandomSampler(idx[split:])
    valid_sampler = SubsetRandomSampler(idx[:split])
    
    # Definisanje kako ce podaci biti ucitavani
    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
    test_loader = DataLoader(test_dataset, batch_size=batch_size, sampler=test_sampler)
    valid_loader = DataLoader(valid_dataset, batch_size=batch_size, sampler=valid_sampler)
    
    return train_loader, test_loader, valid_loader    
```

## Pregled podataka

```python
def show_all_keypoints(image, pred_key_pts=None, gt_key_pts=None):
    plt.imshow(image, cmap='gray')
    if pred_key_pts is not None:
        plt.scatter(pred_key_pts[:, 0], pred_key_pts[:, 1], s=30, marker='.', c='r')
  
    if gt_key_pts is not None:
        plt.scatter(gt_key_pts[:, 0], gt_key_pts[:, 1], s=30, marker='.', c='lightgreen')
```

```python
def visualize_output(images, key_pts=None, gt_pts=None, batch_size=10):
    plt.figure(figsize=(25,15))
    for i in range(10):    
        ax = plt.subplot(3, 5, i+1)

        # de-transformacija slika
        image = images[i].data   
        image = image.numpy()   
        image = np.transpose(image, (1, 2, 0))

        # de-transformacija predvidjenih kljucnih tacaka
        pred_key_pts = None
        if key_pts is not None:
            pred_key_pts = key_pts[i].data
            pred_key_pts = pred_key_pts.numpy()
            pred_key_pts = pred_key_pts*(image.shape[0]/4)+image.shape[0]/2

        gt_key_pts = None
        if gt_pts is not None:
            gt_key_pts = gt_pts[i]         
            gt_key_pts = gt_key_pts*(image.shape[0]/4)+image.shape[0]/2
        
        show_all_keypoints(np.squeeze(image), pred_key_pts, gt_key_pts)
        plt.axis('off')
    plt.savefig('images.png')
    plt.show()
```

```python
def sample_output(data_loader, net=None):
    if net is not None:
        net.eval()
        net.to('cpu')
    
    for i, data in enumerate(data_loader):
        images = data['image']
        key_pts = data['keypoints']
        images = images.type(torch.FloatTensor)
        
        output_pts = None
        if net is not None:
            output_pts = net(images)
            output_pts = output_pts.view(output_pts.size()[0], 68, -1)
        
        if i == 0:
            return images, output_pts, key_pts
```

```python
batch_size = 64
image_size = 224
train_loader, test_loader, valid_loader = create_datasets(batch_size,image_size,color=False)
images, _, key_pts = sample_output(test_loader)
```

```python
visualize_output(images, gt_pts=key_pts)
```

# Definisanje modela

```python
import torch.nn as nn
import torch.optim as optim
from torch.optim import lr_scheduler
from torchvision import models

class resnet18_grayscale(nn.Module):
    def __init__(self):
        super().__init__()
        self.resnet18 = models.resnet18(pretrained=True)
        self.resnet18.conv1 = nn.Conv2d(1, 64, kernel_size=(7, 7), stride=(2, 2), padding=(3, 3), bias=False)
        n_inputs = self.resnet18.fc.in_features
        self.resnet18.fc = nn.Linear(n_inputs, 136)
                        
    def forward(self, x):
        x = self.resnet18(x)
        return x
```

```python
model = resnet18_grayscale()
model.cuda()
for param in model.parameters():
    param.requires_grad_(True)

criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
scheduler = lr_scheduler.ReduceLROnPlateau(optimizer, 'min', patience=7, verbose=True)
```

# Treniranje modela

```python
class EarlyStopping:
    def __init__(self,patience=15):
        self.patience = patience
        self.counter = 0
        self.best_score = None
        self.early_stop = False
        self.val_loss_min = np.Inf
    
    def __call__(self, val_loss, model):
        score = -val_loss
        if self.best_score is None:
            self.best_score = score
        elif score < self.best_score:
            self.counter += 1
            print(f'EarlyStopping counter: {self.counter} out of {self.patience}')
            if self.counter >= self.patience:
                self.early_stop = True
        else:
            self.best_score = score
            self.save_checkpoint(val_loss, model)
            self.counter = 0
    
    def save_checkpoint(self, val_loss, model):
        torch.save(model.state_dict(), 'saved_models/checkpoint.pt')
        self.val_loss_min = val_loss        
```

```python
def validation_loss(valid_loader, model):
    model.eval()
    loss, running_loss = 0.0, 0.0
    for i, data in enumerate(valid_loader):
        images = data['image']
        targets = data['keypoints']
        
        targets = targets.view(targets.size(0), -1)
        
        images = images.type(torch.FloatTensor)
        targets = targets.type(torch.FloatTensor)
        images = images.to(device)
        targets = targets.to(device)
        
        output = model(images)
        loss = criterion(output, targets)
        running_loss += loss.item()
    avg_loss = running_loss / (i+1)
    model.train()
    return avg_loss
```

```python
import csv

def write_list_to_file(list_values, filename):
    with open(filename, "w") as outfile:
        for entries in list_values:
            outfile.write(str(entries))
            outfile.write("\n")
```

```python
device = torch.device('cuda')
n_epochs = 200
model.to(device)
early_stop = EarlyStopping(patience=25)
all_train_loss, all_val_loss = [], []
for epoch in range(n_epochs):
    model.train()
    train_loss, val_loss = 0.0, 0.0
    
    for i, data in enumerate(train_loader):
        images = data['image']
        targets = data['keypoints']
        targets = targets.view(targets.size(0), -1)
        
        images = images.type(torch.FloatTensor)
        targets = targets.type(torch.FloatTensor)
        
        images = images.to(device)
        targets = targets.to(device)
        
        optimizer.zero_grad()
        
        output = model(images)
        loss = criterion(output, targets)
        loss.backward()
        optimizer.step()
        
        train_loss += 1/(i + 1) * (loss.data - train_loss)
    
    val_loss = validation_loss(valid_loader, model)
    scheduler.step(val_loss)
    
    print('Epoch: {}/{} \tTraininig Loss: {:.5f} \tValidation Loss: {:.5f}'.format(epoch+1, n_epochs, train_loss, val_loss))
    all_train_loss.append(train_loss)
    all_val_loss.append(val_loss)
    
    early_stop(val_loss, model)
    if early_stop.early_stop:
        break

model_name = 'resnet18_gray'        
model.load_state_dict(torch.load('saved_models/checkpoint.pt'))
val_loss  = validation_loss(valid_loader, model)
torch.save(model.state_dict(), f'saved_models/{model_name}_epochs{epoch}_bs{batch_size}_vloss{val_loss:.5f}.pt')
print('Finished Training')     
```

```python
face_cascade = cv2.CascadeClassifier('detector_architectures/haarcascade_frontalface_default.xml')
```

```python
def draw_face(img, img2, output_pts):
    
    color = (80,255,80)
    for i in range(len(output_pts)):
        if i not in [16, 21, 26, 30, 35]:
            pt1 = (output_pts[i][0], output_pts[i][1])
            if i in [36, 42]:
                color = (255,220,0)
            elif i == 48:
                  color = (0,0,205)
            if i == 41:
                pt2 = (output_pts[36][0], output_pts[36][1])
            elif i == 47:
                pt2 = (output_pts[42][0], output_pts[42][1])
            elif i == 67:
                pt2 = (output_pts[60][0], output_pts[60][1])
            else:
                pt2 = (output_pts[i+1][0], output_pts[i+1][1])
            cv2.line(img2, pt1, pt2, color, thickness=1, lineType=1, shift=0) 
            cv2.circle(img, pt1, 1, color, -1)
```

```python
def detect_key_pts(img, scale):
    img2 = np.zeros_like(img)
    faces = face_cascade.detectMultiScale(img, 1.2, 2)
    
    if len(faces) == 0:
        return cv2.hconcat([img, img2])
    image_copy = np.copy(img)
    
    for (x,y,w,h) in faces:
        if (y-scale) < 0 and (x-scale) < 0:
            if (y-scale) < (x-scale):
                scale += (y-scale)
            else:
                scale += (x-scale)
        elif (y-scale) < 0 :
            scale += (y-scale)
        elif (x-scale) < 0:
            scale += (x-scale)
        roi = image_copy[y-scale:y+h+scale, x-scale:x+w+scale]
        roi_color = np.copy(roi)
        roi = cv2.cvtColor(roi, cv2.COLOR_RGB2GRAY)
        roi = roi/255.0
        h, w = roi.shape
        shape_before_resize = roi.shape
        roi = cv2.resize(roi, (img_size, img_size))
        shape_after_resize = roi.shape
        scaling_factor = shape_before_resize[0]/shape_after_resize[0]
        roi_color = cv2.resize(roi_color, (img_size, img_size))
        roi_copy = np.copy(roi)
        if(len(roi.shape) == 2):
            roi = roi.reshape(roi.shape[0], roi.shape[1], 1)
        roi = roi.transpose((2, 0, 1))
        roi = torch.from_numpy(roi)
        roi = roi.type(torch.FloatTensor)
        roi.unsqueeze_(0)
        roi.to(device)
        output_pts = model(roi)
        output_pts = output_pts.view(68, -1)
        output_pts = output_pts.detach().numpy()   
        output_pts = output_pts*(roi_copy.shape[0]/4)+roi_copy.shape[0]/2
        for pts in output_pts:
            pts[0] = x-scale+pts[0]*scaling_factor
            pts[1] = y-scale+pts[1]*scaling_factor
       
        draw_face(image_copy, img2, output_pts)
        final = cv2.hconcat([image_copy, img2])
        return final
```

```python
import cv2

def camera_face_detect(save=False):
    cap = cv2.VideoCapture(0)
    out = cv2.VideoWriter('video.avi', cv2.VideoWriter_fourcc(*'XVID'), 20.0, (1280,480))

    while True:
        _, frame = cap.read()
        frame = cv2.flip(frame, 1)
        final = detect_key_pts(frame, 30)
        cv2.imshow('Video', final)
        if save:
            out.write(final)

        key = cv2.waitKey(30)
        if key == 27:
            break
    cap.release()
    cv2.destroyAllWindows()    
```
